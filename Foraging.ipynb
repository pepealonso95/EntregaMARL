```python
import lbforaging
from games.foraging import Foraging
import numpy as np
import time
from agents.random_agent import RandomAgent
from agents.fictitiousplay_t import FictitiousPlay
```

```python
game = Foraging(config=None)
```

```python
def run(game, agents, verbose=False, render=False):

    try:
        game.reset()

        cum_rewards = dict(map(lambda agent: (agent, 0), game.agents))

        if render:
            game.render()
            time.sleep(0.5)

        while not game.done():

            actions = dict(map(lambda agent: (agent, agents[agent].action()), game.agents))

            game.step(actions)

            for agent in game.agents:
                cum_rewards[agent] += game.reward(agent)

            if verbose:
                for agent in game.agents:
                        print(f"Agent {agent} reward: {game.reward(agent)}")
                        print(f"Agent {agent} observe: {game.observe(agent)}")
                
            if render:
                game.render()
                time.sleep(0.5)
            
        return cum_rewards

    except Exception as e:
        print(f"An error occurred: {e}")
```

```python
fgagents = dict(map(lambda agent: (agent, FictitiousPlay(game=game, agent=agent)), game.agents))
```

```python
run(game, fgagents, verbose=True, render=False)
```
